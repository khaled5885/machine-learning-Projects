```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df=pd.read_csv('BostonHousing.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
corr=df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
x=df.drop(['medv'],axis=1)
y=df['medv']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict1=model.predict(x_train)
```

```python
score_1=metrics.r2_score(y_train,y_predict1)
score_2=metrics.mean_absolute_error(y_train,y_predict1)
print('R squared error',score_1)
print('Mean Absolute Error',score_2)
```

```python
y_predict=model.predict(x_test)
score_1=metrics.r2_score(y_test,y_predict)
score_2=metrics.mean_absolute_error(y_test,y_predict)
print('R squared error',score_1)
print('Mean Absolute Error',score_2)
```

```python
plt.scatter(y_train,y_predict1)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
